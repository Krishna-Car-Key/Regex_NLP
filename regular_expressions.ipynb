# Finding most used word

```python
pattern = re.compile("[a-zA-Z]+")
words = re.findall(pattern, book.lower())
```

```python
words
```

```python
dictionery = {}
for word in words:
    if word in dictionery.keys():
        dictionery[word] = dictionery[word] + 1
    else:
        dictionery[word] = 1
```

```python
dictionery
```

```python
dictionery.items()
```

```python
word_list = [(value, key) for (key, value) in dictionery.items()]
```

```python
sorted(word_list, reverse=True)
```

# storing the chapter names

```python
pattern = re.compile("[A-Za-z ]+\n")
words = re.findall(pattern, book)
```

```python
words = [word.replace("\n", "") for word in words]
```

```python
words
```

```python
def get_occurence(word):
    try :
        pattern = re.compile("[^a-zA-Z].*(%s).*[^a-zA-Z]"%word)
        words = re.findall(pattern, book.lower())
        length = len(words)
    except KeyError:
        length = f"The word '{word}' is not in the book."
        
    return length
        
    
```
